# The average salary analysis

## Year : 2014

```python
df_2014 = df.loc[df['year'] == 2014]


fig, ax = plt.subplots(figsize=(12,12))
rank_cp = df_2014.loc[df_2014['soc_name'] == 'Computer programmers'].avg_wage_rank.values
data = df_2014.loc[abs(df_2014['avg_wage_rank']-rank_cp) <= 4]
data = data.sort_values(by = 'avg_wage')
ind = np.arange(len(data['soc_name']))
rects = ax.barh(ind, np.array(data['avg_wage'].values), color='c')
ax.set_yticks(ind)
ax.set_yticklabels(data['soc_name'].values, rotation='horizontal')
ax.set_xlabel("average salary")
ax.set_title("average salary analysis")
#autolabel(rects)
plt.show()
```

### treemap for workforce in industry

```python
df_treemap_indus = pd.read_csv('data/Data USA - Tree Map of Industries by Share for Computer Programmers.csv')
```

```python
df_treemap_indus_14 = df_treemap_indus.loc[df_treemap_indus['year']==2014]
df_treemap_indus_15 = df_treemap_indus.loc[df_treemap_indus['year']==2015]
df_treemap_indus_16 = df_treemap_indus.loc[df_treemap_indus['year']==2016]
df_treemap_indus_14.drop_duplicates(subset ="naics_name", 
                     keep = 'first', inplace = True)
df_treemap_indus_15.drop_duplicates(subset ="naics_name", 
                     keep = 'first', inplace = True)
df_treemap_indus_16.drop_duplicates(subset ="naics_name", 
                     keep = 'first', inplace = True)
df_treemap_indus_14 = df_treemap_indus_14[['naics_name','num_ppl']]
df_treemap_indus_15 = df_treemap_indus_15[['naics_name','num_ppl']]
df_treemap_indus_16 = df_treemap_indus_16[['naics_name','num_ppl']]
```

```python
import squarify
```

```python
# New dataframe, containing only players with more than 0 goals.
dataGoals = df_treemap_indus_14.sort_values(by = 'num_ppl',ascending=False).iloc[0:10]

#Utilise matplotlib to scale our goal numbers between the min and max, then assign this scale to our values.
norm = matplotlib.colors.Normalize(vmin=min(dataGoals.num_ppl), vmax=max(dataGoals.num_ppl))
#colors = [matplotlib.cm.Blues(norm(value)) for value in dataGoals.num_ppl]

#Create our plot and resize it.
fig = plt.gcf()
ax = fig.add_subplot()
fig.set_size_inches(40, 20)

#Use squarify to plot our data, label it and add colours. We add an alpha layer to ensure black labels show through
squarify.plot(label=dataGoals.naics_name+'\n'+dataGoals.num_ppl.astype(str),sizes=dataGoals.num_ppl, alpha=.8, text_kwargs={'fontsize':20})
plt.title("work force in different industry",fontsize=30,fontweight="bold")

#Remove our axes and display the plot
plt.axis('off')
plt.show()
```

## skill analysis

```python
df_skill_bar = pd.read_csv('data/Data USA - Bar Chart of Skill for Computer Programmers.csv')
```

```python
df_skill_bar
```

```python
import seaborn as sns
labels=df_skill_bar['skill_name'].values
stats=df_skill_bar['value'].values

angles=np.linspace(0, 2*np.pi, len(labels), endpoint=False)
# close the plot
stats=np.concatenate((stats,[stats[0]]))
angles=np.concatenate((angles,[angles[0]]))

fig=plt.figure(figsize=(20,20))
ax = fig.add_subplot(111, polar=True)
ax.plot(angles, stats, 'o-', linewidth=2)
ax.fill(angles, stats, alpha=0.25)
ax.set_thetagrids(angles * 180/np.pi)
plt.xticks(angles, labels, rotation=30,fontsize=15)
ax.set_title('skill',fontsize=30)
ax.grid(True)
```

```python
fig, ax = plt.subplots(figsize=(20,20))
data = df_skill_bar.sort_values(by = 'value')
ind = np.arange(len(data['value']))
rects = ax.barh(ind, np.array(data['value'].values), color='b')
ax.set_yticks(ind)
ax.set_yticklabels(data['skill_name'].values, rotation='horizontal', fontsize =20)
ax.set_xlabel("value",fontsize=30)
ax.set_title("Most inportant skills for computer programmer", fontsize= 30)
#autolabel(rects)
plt.show()
```
